```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from plotting_helpers import *
```

```python
#get data of interest
random_data = pd.read_csv("random_edges/25_2_100_vi.csv")
pruned_data = pd.read_csv("pruned_edges/25_2_10_vi.csv")
```

```python
#get summary stats data
random_avg, random_std = perf_summary_stats(random_data)
pruned_avg, pruned_std = perf_summary_stats(pruned_data)
```

```python
#draw some basic perf plots
plt.plot(random_avg.index, random_avg, label = "Random Edges")
plt.plot(pruned_avg.index, pruned_avg, label = "Pruned Edges")
plt.xlabel("Number of edges")
plt.ylabel("Variation of Information")
plt.title("Performance differences with 25 nodes and 2 stored states")
plt.legend();
```

Right now have a clear idea that pruned edges will always do better than random edges for the middling number of edges. Think an interesting question is how does this relationship change as we change the number of stored states. Right now we've only considered 25 nodes with 2 stored states, but let's see what happens when we change that to 1 to 4 stored states (25 * .15 = 3.75 expected number of possible stored states with good perf).

```python
#same procedure as above with 25 nodes and 1 stored state

#get data of interest
random_data_1stored = pd.read_csv("random_edges/25_1_10_vi.csv")
pruned_data_1stored = pd.read_csv("pruned_edges/25_1_10_vi.csv")

#get summary stats data
random_avg_1stored, random_std_1stored = perf_summary_stats(random_data_1stored)
pruned_avg_1stored, pruned_std_1stored = perf_summary_stats(pruned_data_1stored)

plt.plot(random_avg_1stored.index, random_avg_1stored, label = "Random Edges")
plt.plot(pruned_avg_1stored.index, pruned_avg_1stored, label = "Pruned Edges")
plt.xlabel("Number of edges")
plt.ylabel("Variation of Information")
plt.title("Performance differences with 25 nodes and 1 stored state")
plt.legend();
```
